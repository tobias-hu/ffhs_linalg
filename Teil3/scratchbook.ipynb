```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4 * np.pi, 4 * np.pi, 1000)

y = np.cos(x)

plt.plot(x, y)

plt.title('Kosinus-Funktion')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(0, np.pi, 1000)

y = np.cos(x)

plt.plot(x, y)

plt.title('Kosinus-Funktion')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='black')
plt.axvline(0, color='black')

x_stuetz = [math.pi/6, math.pi/2, 5*math.pi/6]
y_stuetz = np.cos(x_stuetz)

plt.scatter(x_stuetz, y_stuetz, color='red')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(0, np.pi, 1000)

y = np.full_like(x, np.cos(0))

plt.plot(x, y)

plt.title('Kosinus-Funktion')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='black')
plt.axvline(0, color='black')

x_stuetz = [math.pi/6, math.pi/2, 5*math.pi/6]
y_stuetz = [1,1,1]

plt.scatter(x_stuetz, y_stuetz, color='red')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(0, np.pi, 1000)

y = np.cos(2*x)

plt.plot(x, y)

plt.title('Kosinus-Funktion')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='black')
plt.axvline(0, color='black')

x_stuetz = [math.pi/6, math.pi/2, 5*math.pi/6]
y_stuetz = np.cos(2*np.array(x_stuetz))

plt.scatter(x_stuetz, y_stuetz, color='red')

plt.show()
```
